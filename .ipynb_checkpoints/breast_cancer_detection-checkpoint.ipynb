```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr("spearman")
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features_mean = correlation(df_mean, 0.85)
corr_features_se = correlation(df_se, 0.85)
corr_features_worst = correlation(df_worst, 0.85)

display(corr_features_mean, corr_features_se, corr_features_worst)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df["diagnosis"] = encoder.fit_transform(df['diagnosis'])# encode malignant as 1 and benign as 0 
```

# Feature Selection

```python
display(corr_features_mean, corr_features_se, corr_features_worst)
coldrop=['fractal_dimension_mean',
 'texture_se',
 'smoothness_se',
 'symmetry_se',
 'fractal_dimension_se']

corr_features_mean = list(corr_features_mean)
corr_features_se = list(corr_features_se)
corr_features_worst = list(corr_features_worst)

list_drop = corr_features_mean + corr_features_se + corr_features_worst + coldrop
list_drop = list(set(list_drop))
list_drop.append("id")
list_drop
```

Dropping these features as some of them are highly correlated with independent features and hence acting as duplicates and some don't have any relation with target feature.

```python
df.drop(list_drop, axis=1, inplace=True)
display(df.columns)
display(df.shape)
```

```python
df.head()
```

```python
# target feature separation
x = df.drop("diagnosis", axis=1)
y = df.diagnosis
```

# Train Test|split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)  #70 percent training 30 percent testing
```

```python
y_train.shape, y_train[y_train==1].shape, y_train[y_train==0].shape
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dt = DecisionTreeClassifier(max_depth=3)
```

```python
model_dt.fit(x_train,y_train)
```

```python
base_pred=model_dt.predict(x_test)
```

# Evaluation

```python
confusion_matrix(y_test,base_pred)
```

```python
plot_confusion_matrix(model_dt,x_test,y_test)
```

```python
print(classification_report(y_test,base_pred))
print("Accuracy:",metrics.accuracy_score(y_test, base_pred))
print("Precision:",metrics.precision_score(y_test, base_pred))
print("Recall:",metrics.recall_score(y_test, base_pred))
dt_train_acc = model_dt.score(x_train,y_train)
dt_test_accuracy = model_dt.score(x_test,y_test)
```

# Comparison

```python
def comp_model(model_list, model_train_acc_list, model_test_acc_list):
    data = {"Model Name": model_list, "Train Accuracy(%)": [i*100 for i in model_train_acc_list], "Test Accuracy(%)": [i*100 for i in model_test_acc_list]}
    Comparision = pd.DataFrame(data)
    return Comparision
```

```python
model_list = ["Decision Tree Classifier", "SVM", "Random Forest Classifier"]
model_train_acc_list = [dt_train_acc, svm_train_acc, rf_train_acc]
model_test_acc_list = [dt_test_accuracy, svm_test_accuracy, rf_test_accuracy]
comp_model(model_list, model_train_acc_list, model_test_acc_list)
```
